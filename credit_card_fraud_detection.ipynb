```python
import numpy as np
import pandas as pd
from sklearn.manifold import TSNE
import sklearn
import kmapper as km
import shap
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/home/sid/Codes/data/creditcardfraud/creditcard.csv')
df.columns
```

```python
from sklearn.preprocessing import RobustScaler
rob_scaler = RobustScaler()

df['s_amount'] = rob_scaler.fit_transform(df['Amount'].values.reshape(-1,1))
df['s_time'] = rob_scaler.fit_transform(df['Time'].values.reshape(-1,1))

df.drop(['Time','Amount'],axis=1,inplace=True)

df = df.sample(frac=1)
df.shape
```

```python
fraud_df = df.loc[df['Class']==1]
safe_df = df.loc[df['Class']==0]
normal_dist_df = pd.concat([fraud_df,safe_df])
new_df = normal_dist_df.sample(frac=1,random_state=42)

x = new_df.drop('Class',axis=1)
y = new_df['Class']

x = np.array(x)
y = np.array(y)
#y = np.array([1 if z==0 else 0 for z in y])
```

```python
mapper = km.KeplerMapper(verbose=4)
x.shape

x_reduced = TSNE(n_components=2).fit_transform(x)
#x_reduced = sklearn.decomposition.PCA(n_components=2).fit_transform(x)
f,ax = plt.subplots(1,1,figsize=(20,30))
ax.scatter(x_reduced[:,0],x_reduced[:,1],c=(y==0),cmap='coolwarm',label='Safe',linewidths=2)
ax.scatter(x_reduced[:,0],x_reduced[:,1],c=(y==1),cmap='coolwarm',label='Fraud',linewidths=2)
ax.grid(True)
plt.show()
```

```python
projected_data = mapper.fit_transform(x,projection=TSNE())
```

```python
html_path = '/home/sid/Codes'
metrics = ['haversine','cosine','euclidean']

for metric in metrics:
    graph = mapper.map(projected_data,
                      clusterer=sklearn.cluster.DBSCAN(eps=50,min_samples=5,metric=metric),
                      cover=km.Cover(40,0.3))
    mapper.visualize(graph,title='Credit Card Mapper',path_html=html_path+'/map_'+str(metric)+'.html',color_function=y,custom_tooltips=y)
    km.draw_matplotlib(graph)
    plt.show()
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.linear_model import LogisticRegression

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
classifier = LogisticRegression()
classifier.fit(x_train,y_train)
train_score = cross_val_score(classifier,x_train,y_train,cv=5)
print('Train score:',train_score)

best_params = {"penalty": ['l1', 'l2'], 'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
grid_logreg = GridSearchCV(LogisticRegression(),best_params)
grid_logreg.fit(x_train,y_train)
logreg = grid_logreg.best_estimator_
logreg_score = cross_val_score(logreg,x_train,y_train,cv=5)
print('Log score:',logreg_score)
```

```python
from sklearn.metrics import confusion_matrix

y_pred = classifier.predict(x_test)
conf = confusion_matrix(y_test,y_pred)
print(conf)
```
